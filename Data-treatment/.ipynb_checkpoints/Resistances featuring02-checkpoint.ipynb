```python
import pandas as pd
from scipy.optimize import curve_fit
from sklearn.cluster import KMeans
from scipy.interpolate import interp1d
import seaborn as sb
```

```python
def func(x, a, b, c):
    return a * np.exp(-b * x) + c

def func(x):
    return 10
```

```python
Din = pd.read_csv('DATA/50W/D20190212e.txt',
                 names=['Time','Temperature','RH','Voltage','Current'])
#fig, ax1 = plt.subplots(figsize=(40,10))
fig, ax1 = plt.subplots()
ax1.scatter(Din.Time/1000,Din.Temperature, marker = '.', c= 'orange')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Temperature (Â°C)', color = 'orange')
ax1.tick_params('y', colors = 'orange')
#ax1.axis([0,700,25.5,37])
ax1t = ax1.twinx()
ax1t.scatter(Din.Time/1000,Din.RH, marker = '.',c='b')
ax1t.set_ylabel('Relative humidity (%)', color = 'blue')
ax1t.tick_params('y', colors = 'blue')
#ax1t.axis([0,700,21,42])
ax1.grid(True)
#ax1.axis([0,100,22,50])
#plt.tight_layoutut()
plt.show()
```

```python
Din.Temperature.describe()
```

```python
plt.scatter(Din.Temperature,Din.RH,
         color = 'green', marker = '.')
popt, pcov = curve_fit(func, Din.Temperature, Din.RH)
plt.plot(Din.Temperature, func(Din.Temperature, *popt), color = 'red',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
plt.xlabel('Temperacture (Â°C)')
plt.ylabel('Relative humidity (%)')
plt.title('Relative humidty vs. temperature')
plt.legend()
```

#### Interpolation

```python
x = np.asarray(Din.Temperature)
y = np.asarray(Din.RH)
o = np.transpose(np.array([x,y]))
o = np.sort(o, axis = 0)
f = interp1d(o[:,0], o[:,1])
f2 = interp1d(o[:,0], o[:,1], kind='cubic')
```
